```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# criando array de salario

salario = np.array([100, 200,300,500,400,150])

salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c="Black")
```

```python
plt.plot(salario, c="Black", ls="--")
```

```python
plt.plot(salario, ls='--', marker='s')
```

```python
plt.plot(salario, c='Black', ls='--', marker='s', ms=10)
plt.show()
```
